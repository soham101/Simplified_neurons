```
import numpy as np
import pandas as pd
import matplotlib.pyplot as mlt
import tensorflow as tf

sess=tf.Session()

np.random.seed(1)
```


##Neural network of 2 hidden layers n_x=4, n_h1=50,  n_h2=3, n_y=3.
      LINEAR->RELU, LINEAR->RELU, LINEAR->SOFTMAX


```
def initialize_parameters(n_x, n_h1, n_h2, n_y):
  
  
  parameters={}
  parameters['W1']=np.random.randn(n_h1, n_x)*0.01
  parameters['b1']=np.zeros((n_h1, 1))
  parameters['W2']=np.random.randn(n_h2, n_h1)*0.01
  parameters['b2']=np.zeros((n_h2, 1))
  parameters['W3']=np.random.randn(n_y, n_h2)*0.01
  parameters['b3']=np.zeros((n_y, 1))
  
  assert parameters['W1'].shape == (n_h1, n_x)
  assert parameters['b1'].shape == (n_h1, 1)
  assert parameters['W2'].shape == (n_h2, n_h1)
  assert parameters['b2'].shape == (n_h2, 1)
  assert parameters['W3'].shape == (n_y, n_h2)
  assert parameters['b3'].shape == (n_y, 1)
         
  return parameters
```

```
def relu_deriv(x):
    x[x>=0]=1
    x[x<=0]=0
    return x
```

```
def activation(a_prev, w, b, funct):
    
    #print(b.shape)
    z=np.dot(w, a_prev)+b
   
    
    if funct=="relu":
      a=tf.nn.relu(z, name=None)
    elif funct=="sigmoid":
      a=tf.sigmoid(z, name=None)
    elif funct=="softmax":
      a=tf.nn.softmax(z.T, name=None)
      
      
      a=sess.run(a)
      
    return a, z
```

```
def forw_prop(x_train, parameters):
    
    cache={}
    
    cache['a1'], cache['z1']=activation(x_train, parameters['W1'], parameters['b1'], funct="relu")
    with sess.as_default():
      assert tf.get_default_session() is sess
      cache['a1']=cache['a1'].eval()
    cache['a2'], cache['z2']=activation(cache['a1'], parameters['W2'], parameters['b2'], funct="relu")
    with sess.as_default():
      assert tf.get_default_session() is sess
      cache['a2']=cache['a2'].eval()
    cache['a3'], cache['z3']=activation(cache['a2'], parameters['W3'], parameters['b3'], funct="softmax")
    
    return cache
```

```
def back_prop(x_train, y_train, cache, parameters):
    grads={}
    m = y_train.shape[1]
    cache['a3']
    
    dZ3=cache['a3'].T-y_train
    grads['dW3']=np.dot(dZ3, cache['a2'].T)/m
    grads['db3']=np.sum(dZ3, axis=1, keepdims=True)
    dZ2=np.multiply(np.dot(parameters['W3'].T, dZ3), relu_deriv(cache['z2']))
    grads['dW2']=np.dot(dZ2, cache['a1'].T)/m
    grads['db2']=np.sum(dZ2, axis=1, keepdims=True)
    dZ1=np.multiply(np.dot(parameters['W2'].T, dZ2), relu_deriv(cache['z1']))
    grads['dW1']=np.dot(dZ1, x_train.T)/m
    grads['db1']=np.sum(dZ1, axis=1, keepdims=True)
    #grads['db1'].shape
    
    grads['db3']=np.array([grads['db3']])
    
    return grads
```

```
def cost(a3, y_train):
    m = y_train.shape[1]
    
    y_train=y_train.T
    #print(y_train.shape)
    a3=a3.T
    #print(a3)
    cost = -np.sum(np.add(np.add(np.multiply(y_train[0], np.log(a3[0])), np.multiply((y_train[1]), np.log(a3[1]))), np.multiply((y_train[2]), np.log(a3[2]))))/m
    
    cost = np.squeeze(cost)      
    assert(cost.shape == ())
    #print(np.multiply(y_train[0], np.log(a3[0]))-np.multiply((1-y_train[0]), np.log(1-a3[0])))
    
    return cost
```

```
df=pd.read_csv("/content/gdrive/My Drive/ML/Iris.csv")
X=df.drop(['Species','Id'], axis=1)
Y=df['Species']
Y=np.array([Y])
Y=pd.get_dummies(Y[0])

#print(Y.shape)

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(X, Y, test_size=0.35, random_state=10)
x_train=x_train.T
x_test=x_test.T
y_train=y_train.T
y_test=y_test.T    
  
parameters=initialize_parameters(4, 50, 3, 3)
cache=forw_prop(x_train, parameters)
a3=cache['a3']
cst=cost(a3, y_train)
#print(cost)
epochs=500
lr=0.01
costs=[]
itr=[]

for i in range(1, epochs):
 


  def cost(a3, y_train):
    m = y_train.shape[1]
    
    y_train=y_train.T
    #print(y_train.shape)
    a3=a3.T
    #print(a3)
    cost = -np.sum(np.add(np.add(np.multiply(y_train[0], np.log(a3[0])), np.multiply((y_train[1]), np.log(a3[1]))), np.multiply((y_train[2]), np.log(a3[2]))))/m
    
    cost = np.squeeze(cost)      
    assert(cost.shape == ())
    #print(np.multiply(y_train[0], np.log(a3[0]))-np.multiply((1-y_train[0]), np.log(1-a3[0])))
    
    return cost
  
  itr.append(i)
  
  cache=forw_prop(x_train, parameters)
  cst=cost(a3, y_train)
  grads=back_prop(x_train, y_train, cache, parameters)
  
  parameters['W3']=parameters['W3']-lr*grads['dW3']
  parameters['b3']=parameters['b3']-lr*grads['db3']
  parameters['W2']=parameters['W2']-lr*grads['dW2']
  parameters['b2']=parameters['b2']-lr*grads['db2']
  parameters['W1']=parameters['W1']-lr*grads['dW1']
  parameters['b1']=parameters['b1']-lr*grads['db1']
  
  
  #print(parameters['b1'])
  
  print("epochs:= "+str(i)+", the cost:= "+str(cst))
  
  costs.append(cst)
```

```
mlt.plot(itr,costs)
mlt.show()
```
